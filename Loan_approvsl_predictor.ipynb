```python

df = pd.read_csv('loan_approval_dataset.csv')
df
```

```python
df.drop(columns = ['loan_id'], inplace=True)
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df['Assets'] = df.residential_assets_value + df.commercial_assets_value + df.luxury_assets_value + df.bank_asset_value
```

```python
df.drop(columns = ['residential_assets_value', 'commercial_assets_value', 'luxury_assets_value', 'bank_asset_value'], inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.education.unique()
```

```python
def clean_data(st):
  st = st.strip()
  return st
```

```python
clean_data('Graduate')
```

```python
df.education = df.education.apply(clean_data)
```

```python
df.education.unique()
```

```python
df['education'] = df['education'].replace(['Graduate', 'Not Graduate'],[1, 0])
```

```python
df.self_employed.unique()
```

```python
df.self_employed = df.self_employed.apply(clean_data)
```

```python
df.self_employed.unique()
```

```python
df.self_employed = df.self_employed.replace(['Yes', 'No'],[1, 0])
```

```python
df.loan_status.unique()
```

```python
df.loan_status = df.loan_status.apply(clean_data)
```

```python
df.loan_status.unique()
```

```python
df.loan_status = df.loan_status.replace(['Approved', 'Rejected'], [1, 0])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_df = df.drop(columns=['loan_status'])
output_df = df['loan_status']
```

```python
input_df
```

```python
output_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_df, output_df, test_size=0.2)
```

```python
x_train.shape, x_train.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_scaled, y_train)
```

```python
model.score(x_test_scaled, y_test)
```

```python
pred_data = pd.DataFrame([['2',	'1',	'0',	'9600000',	'29900000',	'12',	'778',	'50700000']],columns =['no_of_dependents', 'education', 'self_employed', 'income_annum',
       'loan_amount', 'loan_term', 'cibil_score', 'Assets'])
```

```python
pred_data = scaler.transform(pred_data)
```

```python
model.predict(pred_data)
```
